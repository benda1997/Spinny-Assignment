```python
import pandas as pd
```

```python
df=pd.read_csv('Data_2.csv')
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df.columns
```

```python
#Feature Imp For Overall
features = ['no_of_dependents', 'income_annum', 'loan_amount',
        'loan_term', 'residential_assets_value',
        'commercial_assets_value', 'luxury_assets_value', 'bank_asset_value',
        'education_ranking', 'employment_type_Business',
        'employment_type_Freelancer', 'employment_type_Salaried', 'loan_range',
        'income_range', 'Loan_by_Income', 'Income_perDependent',
        'Loan_by_Income_perDependent', 'Loan_Per_Year',
        'Loan_Per_Year_by_income_annum', 'Loan_Per_Year_by_Income_perDependent',
        'Total_Assets', 'Real_Estate_Assests', 'Loan_by_Assests',
        'Total_Income_Loan_Duration',
        'Loan_by_Total_income', 'Loan_by_Assets_Total_income',
        'Loan_by_RealEstateAssets_Total_income',
        'Loan_by_RealEstateAssets_BankAssets_Total_income',
        'Loan_by_TotalIncomeAsset_per_dependent',
        'Luxury_assets_by_Total_assets', 'Liquidity_Ratio','cibil_score','cibil_range','Loan_approval_Pred_based_Cibil'
       ]
target = 'loan_status'

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(df[features], df[target])

feature_importances = rf_classifier.feature_importances_

feature_importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
# Display the feature importances
print(feature_importance_df)

import matplotlib.pyplot as plt

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # Invert y-axis to display most important features at the top
plt.show()
```
